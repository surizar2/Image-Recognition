```python
import cv2
from google.colab.patches import cv2_imshow

# Ruta de la imagen del jardn
image_path = 'jardin1.jpeg'

# Cargar la imagen
image = cv2.imread(image_path)
image_og = cv2.imread(image_path)

# Convertir la imagen a escala de grises
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Aplicar un filtro Gaussiano para reducir el ruido
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detectar los bordes en la imagen utilizando Canny
edges = cv2.Canny(blurred, threshold1=30, threshold2=100)

# Encontrar contornos en la imagen
contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Contar el nmero de tallos de csped
num_tallos = len(contours)

# Dibujar los contornos en la imagen original
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)

# Mostrar el resultado
cv2_imshow(image)

# Imprimir el nmero de tallos de csped encontrados
print(f'Nmero de tallos de csped: {num_tallos}')
cv2_imshow(image_og)
```
